```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
X = dataset.iloc[:, 3:].values
```

```python
from sklearn.cluster import KMeans
help(KMeans)
```

```python
wcss = []
for i in range(1,11):
    kmean = KMeans(n_clusters=i, init='k-means++', n_init=10, max_iter=300, random_state=0)
    kmean.fit(X)
    wcss.append(kmean.inertia_)
plt.plot(range(1,11), wcss)
plt.show()
```

```python
kmean = KMeans(n_clusters=6, init='k-means++', random_state=0)
y_pred = kmean.fit_predict(X)
print(y_pred)
```

```python
plt.scatter(X[y_pred==0,0], X[y_pred==0,1], c='red', label='cluster 1')
plt.scatter(X[y_pred==1,0], X[y_pred==1,1], c='blue', label='cluster 2')
plt.scatter(X[y_pred==2,0], X[y_pred==2,1], c='green', label='cluster 3')
plt.scatter(X[y_pred==3,0], X[y_pred==3,1], c='pink', label='cluster 4')
plt.scatter(X[y_pred==4,0], X[y_pred==4,1], c='cyan', label='cluster 5')
plt.scatter(X[y_pred==5,0], X[y_pred==5,1], c='black', label='cluster 5')
plt.scatter(kmean.cluster_centers_[:, 0], kmean.cluster_centers_[:, 1], c = 'yellow', label = 'Centroids')
plt.title('Clusters of customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
plt.scatter(X[:,0],X[:,1])
plt.show()
```
